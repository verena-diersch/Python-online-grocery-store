### 05. Producing crosstabs for different indicators like family status, departments or region

```python
# Counting orders by family status
df_final['family_status'].value_counts()
```

```python
# Crossing orders by department and family status
crosstab_family_status_dept = pd.crosstab(df_final['department'], df_final['family_status'], dropna = False)
```

```python
crosstab_family_status_dept
```

```python
# Crossing orders by region and family status
crosstab_family_status_region = pd.crosstab(df_final['family_status'], df_final['region'], dropna = False)
```

```python
crosstab_family_status_region
```

```python
# Crossing age_range and family status
crosstab_family_status_age = pd.crosstab(df_final['family_status'], df_final['age_group'], dropna = False)
```

### 03. If-statements with user-defined functions

```python
# Creating a subset with the first one million rows
df = df_ords_prods_merged[:1000000]
```

```python
# Defining the price-label-function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High-range product'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Checking the most expensive product
df['prices'].max()
```

### 04. If-statements with the loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Using the loc function on the entire data frame
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

### 05. If-statements with for-loops

```python
# Printing the frequency of the order_dow column
df_ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
# Counting the busiest, least busy and regular busy day of the week

result = []

for value in df_ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merged['busiest day'] = result
```

```python
df_ords_prods_merged['busiest day'].value_counts(dropna = False)
```

### 06. Task

```python
# Calculating the two busiest and two slowest days
result2 = []

for value in df_ords_prods_merged["order_dow"]:
  if value <= 1: # 0, 1 Saturday and Sunday 
    result2.append("Busiest days")
  elif value >= 3 and value <=4: # 3, 4 Tuesday and Wednesday 
    result2.append("Slowest days")  
  else: #2, 5, 6, Monday, Thursday, Friday
    result2.append("Regularly busy")
```

```python
result2
```

```python
df_ords_prods_merged['busiest days'] = result2
```

```python
# Check the values of this new column for accuracy
df_ords_prods_merged['busiest days'].value_counts(dropna = False)
```

Saturday (0) and Sunday (1), the busiest days, represent 36,5% of the total orders. 
Tuesday (3) and Wednesday (4) are the slowest days which represent 23,5% of the total orders.
Monday (2), Thursday (5) and Friday (6), the regular busy days, represent almost 40% of the total orders.

```python
# Finding and counting the busiest hour of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Finding, counting and grouping the busiest hour of the day

# Counts in millions: hours 10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19 as Most Orders
# Counts in hundreds of thousands: hours 20, 7, 21, 22, 23, 6, 0, 1 as Average Orders
# Counts in tens of thousands: hours 5, 2, 4, 3 as Fewest Orders


df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19]),'busiest_period_of_day')] = 'Most Orders'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([20, 7, 21, 22, 23, 6, 0, 1]),'busiest_period_of_day')] = 'Average Orders'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['order_hour_of_day'].isin([5, 2, 4, 3]),'busiest_period_of_day')] = 'Fewest Orders'
```

```python
# Printing the frequency of the busiest_period_of_day column
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
